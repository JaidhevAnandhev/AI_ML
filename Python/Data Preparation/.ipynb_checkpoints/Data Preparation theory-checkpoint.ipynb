* ``` Any data that are greater than 3 are considered to be outliers.```

````python
# First take the backup of the original data
# It is used to compare the data with original data
dfp['referral_current_salary_1'] = dfp.referral_current_salary
dfp['referral_current_salary_2'] = dfp.referral_current_salary
````

# Method 1. Z-Score
* ``` Formula for Z-Score is X-X̅ / S.D (Standard deviation).```
* ``` X̅ is mean and std() is the (Standard deviation).```

````python
dfp['referral_current_salary_zscore'] = (dfp.referral_current_salary - dfp.referral_current_salary.mean())/dfp.referral_current_salary.std()
dfp['referral_current_salary_zscore'].describe()
````

````python
# To check any outliers are present in the data
dfp[dfp['referral_current_salary_zscore'] > 3] 
````

````python
dfp[dfp['referral_current_salary_zscore'] > 3].referral_current_salary
````

````python
# Let's calculate the value that where we need to cap the outliers
referral_current_sal_impute = (3 * dfp.referral_current_salary.std()) + dfp.referral_current_salary.mean()
````

````python
referral_current_sal_impute
````

````python
round(referral_current_sal_impute, 2)
# thus with this value we can avoid the problem of outliers
````

````python
list1 = dfp[dfp['referral_current_salary_zscore'] > 3].index
list1
````

````python
# Now we have to remove this outlier of our data
dfp['referral_current_salary'] = np.where(dfp.index.isin(list1), round(referral_current_sal_impute, 2), dfp['referral_current_salary'])
````

````python
# Creating a function to detect the range of the boxplot for our column an to detect the outliers
def detect_outlier(col):
        Q1,Q3 = np.percentile(col, [25, 75])
        IQR = Q3 - Q1
        lower_range = Q1-(1.5 * IQR)
        upper_range = Q3+(1.5 * IQR)
        return lower_range, upper_range
````

````python
lr, ur = detect_outlier(dfp['referral_current_salary_1'])
print("Lower range = ", lr, "Upper Range = ", ur)
````

````python
# Since this value of upper an dlower ranges are not there in our data
# hence we need to remove treat this data thus we can remove the outlier

# Replacing the upper range with 247.5
dfp['referral_current_salary_1'] = np.where(dfp.referral_current_salary_1 > ur, ur, dfp.referral_current_salary_1)

# Repalcing the lower range with -76.5
dfp['referral_current_salary_1'] = np.where(dfp.referral_current_salary_1 < lr, lr, dfp.referral_current_salary_1)
````
