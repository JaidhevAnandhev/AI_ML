#  DATA VISUALIZATION WITH EDA

## Data analysis with visualization includes 
* ```Univartiate analysis.```
* ``` Bivartiate analysis.```
* ``` Multvariate analysis.```

# 1. Unvariate analysis
  ``` In univariate analysis we take only one field at a time and analyze the data visually or non-visually and try to seek insights from the data.```

````python
# impot the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
import seaborn as sb
````

````python
dfv = pd.read_csv('Referral_Join_Prediction.csv')
````

````python
dfv
````

````python
dfv[['referral_exp_in_years', 'referral_current_salary']].describe()
````

##  Now the visualization of the data is done using univariate analysis

## 1. Numerical data Univariate analysis

````python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (10,8))
sb.histplot(dfv['referral_exp_in_years'], kde = True, ax = axes[0][0])
sb.boxplot(dfv['referral_exp_in_years'], data=dfv, ax = axes[0][1])
sb.histplot(dfv['referral_current_salary'], kde = True, ax = axes[1][0])
sb.boxplot(dfv['referral_current_salary'], data=dfv, ax = axes[1][1])
plt.show()
````

## 2. Categorical data univariate analysis

### 1) Non graphical data 

````python
dfv['referral_preferred_city'].value_counts()
````

### 3) Graphical data

````python
sb.countplot(x='referral_preferred_city', data=dfv, palette='pastel')
````

# Bivariate analysis

# Numeric Vs Numeric 

````python
# Numeric Vs Numeric
sb.scatterplot(x='referral_exp_in_years', y = 'referral_current_salary', data=dfv)
````

````python
dfv[dfv['referral_exp_in_years']>40]
````

# Categorical Value 

````python
sb.countplot(data=dfv, x='Emp_Salary_Grade', hue='Joined')
````

````python
pd.crosstab(dfv['Emp_Salary_Grade'], dfv['Joined'], margins=True, normalize=True)
# Normalize returns the percentage instead of count
````

## Categorical Vs Numerical Data Analysis

````python
fig_dims = (12,12)
sb.boxplot(x='Joined', y='referral_expected_sal', data=dfv)
````

````python
sb.pairplot(dfv)
````

### Heatmap wihich is a function of seaborn returns the relationship of two numerical data in a collured form

````python
plt.figure(figsize=(10,8))
# corr is used to return the correlation of the two numerical data
# By default annot is FALSE so set it to TRUE
# fmt parameter allows to add string (text) values on the cell
# More the light colour more the relationsthip between two variables or features
sb.heatmap(dfv.corr(),annot=True, fmt='.2f')
````

# MULTIVARIATE ANALYSIS
* ``` In this type of analysis we provide more than 2 variables and thus checking the the relation between thew provided variables.```

````python
# In boxplot hue is the colour representation of the data
plt.figure(figsize=(8.3,4))
sb.boxplot(x='Joined', y='referral_age', hue='Emp_Salary_Grade', data=dfv)
````

````python
g = sb.FacetGrid(dfv, col='referral_preferred_city', hue='Joined', col_wrap=3, height=3)
g = g.map(plt.scatter, 'referral_graduation_marks', 'referral_expected_sal')
g.add_legend()
````
