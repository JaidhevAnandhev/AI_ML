# Data Visulaization in python

##### * In python there are various libraries available for visualizing the data in graphical form. 

##### * Some of the widely used Data Visualiztions are Matplotlib, Scilab and Plotly.

````python
# First import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
````

# LINE PLOT
* ```For line plot we use linspace()```

````python
# Simplest ways to plot a line
#Line Plot

#data
X = np.linspace(1,20,100)
Y = np.exp(X)

#Line plot
plt.plot(X, Y)

# Display the plot
plt.show()
````

## Representing the sine wave
* ``` For this we use sin() function.```

````python
X = np.linspace(1,20,100)
Y = np.sin(X)

plt.plot(X,Y)
plt.show()
````

# We can also represent the line by symbols too
* ``` 'b^' This tells that the each points on line should be like ^ with colour blue.```

````python
X = np.linspace(1,20,100)
Y = np.exp(X)

plt.plot(X,Y,'b^')
plt.show()
````

# Multiple line plots in a single graph

````python
# Data
X = np.linspace(5,20,300)
Y1 = X
Y2 = np.square(X)
Y3 = np.sqrt(X)

# Multiple plots in a single graph
plt.plot(X,Y1,'r', X,Y2,'g', X,Y3,'b')

#Display the Graph
plt.show()
````

# SCATTER PLOT
* ``` For this we use scatter() function.```

````python
# To do so first import the CSV file
file1 = pd.read_csv('tips.csv')
````

````python
file1
````

# Now for the above table we now create a scatter plot

````python
# Define X and Y
X = file1['total_bill']
Y = file1['tip']

# Create a sctter typed graph
plt.scatter(X,Y)

# Name the X-Axis and Y-Axis
plt.xlabel('TOTAL BILLS')
plt.ylabel('TIPS')

# Now show the scatter plot graph
plt.show()
````

# Displaying the scatter plot with different point sizes
* ``` Larger the value larger the dot.```
* ``` Smaller the value then smaller the dot.```

````python
X = file1['total_bill']
Y = file1['tip']

plt.scatter(X,Y, s = np.array(Y)**1.8, c = 'green', alpha = 1)

plt.title('Scatter Plot')

plt.xlabel('TOTAL BILL')
plt.ylabel('TIPS')

plt.show()
````

# PIE PLOT
* ```For this we use pie() function.```

````python
# Before that learn how to count the values in a particular column
file1.day.value_counts()
````

````python
# Create a pie graph
plt.pie(file1.day.value_counts(), autopct='%.1f%%', radius=1, labels = ['Sat', 'Sun', 'Thu', 'Fri'] )

# Display the Pie graph
plt.show()
````

````python
# Create a pie graph
plt.pie(file1.day.value_counts(), autopct='%.1f%%', radius=1.2, labels = ['Sat', 'Sun', 'Thu', 'Fri'],
       explode = [0,0.2,0,0.5])

# Display the Pie graph
plt.show()
````

# HISTOGRAM
* ``` It gives the statistics of data in form of histogram i.e; it shows for the continous flow of data.```
* ``` We can set n number of groups which we call here as bins.```

````python
#bins are used for knowing the accurate distruction of data throughout the range.
plt.hist(file1['tip'], bins = 12) 

plt.show()
````

# BOX PLOT

````python
plt.boxplot(file1['total_bill'])
plt.text(x = 1.1, y = file1['total_bill'].min(), s = 'min')
plt.text(x = 1.1, y = file1['total_bill'].max(), s = 'max')

plt.title('TOTAL BILLS')
plt.ylabel('Total Bill')

plt.show()
````

# STRIP PLOT
* ``` It is one way similar to the bar graph but since it is graphics intensive so for not showing the  two boring bars it gives the observation given as two strips so we can accurately tell many count and less count for the two strips instead ofsghowing it as a bar.```

* ``` For using this function we will be using Seaborn library.```

````python
sb.stripplot(x = 'time', y = 'tip', data=file1)

plt.show()
````

## For spreading the data in Dinner side we use 'jitter' parameter in stripplot function therefore moving in left and right.

````python
sb.stripplot(x = 'time', y = 'tip', data=file1, jitter=True)

plt.show()
````

# SWARM PLOT
* ``` It is same as the strip plot except that it does not control the size of the bars allows to be big or more based on the number of observations.```

* ```It give the good representation of tips and time.```

````python
sb.swarmplot(x = 'time', y = 'tip', data =  file1)

plt.show()
````

````python
#Adding colurs as as hue by using another variable called day
# hues are used to represent the data with the differnt colours.
sb.swarmplot(y = 'tip', x = 'time', data=file1, hue='day')

plt.show()
````

#  VIOLIN PLOT
* ``` It does the same thing as the swarm plot except that it provides the shape to the graph unlike the swarm plot.```
* ``` It looks smooth and symmetric.```
* ```It helps to give out a clear difference between the parameters.```

````python
sb.violinplot(y = 'tip', x = 'time', data=file1)

plt.show()
````

````python
# We can also preferably show the data of other variables simultaneously.
sb.violinplot(y = 'total_bill', x = 'day', data=file1, hue='sex', split=True)

plt.show()
````

# PAIR PLOTS
* ``` It displays the graph of all varaibles within the single output.```

````python
sb.pairplot(file1)

plt.show()
````

#  DISPTPLOT
* ``` It provides a graph that contains the combination of Histogram and distribution plot.```
* ``` displot( )  is the function used by default.```
* `` It helps in providing the underlying distribution of the bar/distribution graph.``
* `` It is the approximation of the given data.``

````python
sb.displot(file1['total_bill'])

plt.show()
````

#  COUNT PLOT

````python
sb.countplot(data = file1,
             x = 'day',
             hue='time')

plt.show()
````

# HEAT MAP
* ``` Correlations are used to show colours inside a matrix. It means the relation between two variables.```
* ``` It comes from the corr() function that tells for example correlation between tips and total_bill. and so on.```
* ``` So heat map() is used to show the variations using the colour as seen below.```

````python
corr_matrix = file1.corr()  
````

````python
corr_matrix
````

````python
sb.heatmap(corr_matrix, annot=True)

plt.show()
````
