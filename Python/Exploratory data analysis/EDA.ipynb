# PREPROCESSING THE DATA

## TREAT BAD Data
* ``` Use unique( ) function to display all the unique values of the particular column.```
* ``` If we run the code then you see that '?' is given in unique data list and it shows that the data 
in the list is BAD and should be treated.```
* ``` And also we could find that the data type is object but actually they are numbers.```
* ```Hence it does not allows us to find the mean median and conversion of numerical value.```

````python
df['referral_aptitude_numerical_score'].unique()
````

````python
# We can use replace ? to NaN and to make the make the changes permanent using replace() function.

df['referral_aptitude_numerical_score'].replace('?', np.NaN, inplace=True)
df['referral_aptitude_reasoning_score'].replace('?', np.NaN, inplace=True)
````

````python
df['referral_aptitude_reasoning_score'].unique()
````

````python
# Now change the data type of both the features (Columns) from object to float
df['referral_aptitude_numerical_score'] = df['referral_aptitude_numerical_score'].astype('float')
df['referral_aptitude_reasoning_score'] = df['referral_aptitude_reasoning_score'].astype('float')
````

````python
df['referral_aptitude_numerical_score'].unique()
````

````python
df['referral_aptitude_reasoning_score'].unique()
````

````python
# Now if we see the information the data type would have been changed
df.info()
````

````python
# To check whether the marks scored in 10th is less than 35
df[df['referral_10_th_marks'] < 35]
````

````python
# Now we try to set the 10th marks to 12th marks if the marks scored in 10th is equal to -1 or less than 35
# So if the comdition is true then set the 10 th marks to 12th marks else 
df['referral_10_th_marks'] = np.where(df['referral_10_th_marks'] == -1 , df['referral_12th_marks'], df['referral_10_th_marks'])
````

# Now that we have separted the numerical and categorical data as df_num and df_cat

# SIMPLE IMPUTER
* ``` It helps to replace the NULL data with the straterguical data such as mean median or mode.```
* ``` It is been imported from the Scikit.impute Library.```

````python
# First import the needed libraries
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values = np.nan, strategy='median')
````

````python
# Now we ae creating an updated df_num dataframe
# Since during this preocess the data would be in the form of array so it converted back to the dataFrame
df_num = pd.DataFrame(imputer.fit_transform(df_num), columns=df_num.columns)
````

````python
df_num.isnull().sum()
````

````python
df_num.describe().T
````

````python
df_num.head()
````

# Now we can concatenate the numerical and categorical data 

````python
df_new = pd.concat([df_num, df_cat], axis=1, join='inner')
````

````python
df_new
````

````python
df_new.info()
# Thus we have replaced the null values of numerical data by replacing them with the median value.
````
