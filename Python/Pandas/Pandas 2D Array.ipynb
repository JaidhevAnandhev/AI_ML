* ``` We create data frame from a dictionary. ```
   * ``` Here key becomes the columns and values becomes the rows.```

````python
import pandas as pd
````

````python
# Creating a dictionary and print the output as a table using DataFrame
df1 = pd.DataFrame({"Name":['Jai', 'Raj', 'Sam', 'Ram'], "Age" : [10, 12, 14, 18]})

# Printing the table
df1
````

# Data-Frame in built functions
* ``` head( ) = To look up first 5 rows of a data-frame we use this function.```
* ``` tail( ) = To look up last 5 rows of data we use tail function. ```
* ``` shape ( ) = To know the shape of the data-frame i.e; knowing how many rows and columns are             there. ```
* ``` describe ( ) = To know the basic description of the data-frame. ```

````python
# To get the csv file and view it's data we use read_csv function
iris = pd.read_csv('Iris.csv')
````

````python
# Call the object
iris
````

# head( ) function in Pandas
* ``` This reads the first 5 datas in the CSV File. ```

````python
#head() function Shows up first 5 data by default
iris.head()
````

````python
# tail() shows last 5 data of the CSV File
iris.tail()
````

# shape( ) method
* ``` This tells the number of rwos and columns of the CSV File. ```

````python
iris.shape
````

# describe ( ) method
* ``` This shows the basic description of the table. ```
* Things that shows up are
     * max
     * 75%
     * 50%
     * 25%
     * min
     * std
     * mean
     * count

````python
iris.describe()
````

# Extract the particular rows and columns using iloc function
* Now to extract first 3 rows and 2 columns we code .iloc[0:3, 0:2]
* 0.3 corresponds to rows and 0.2 corresponds to colyumns

````python
iris.iloc[0:5, 0:4]
````

````python
iris.iloc[8:10, 2:4]
````

# To print only the particular column and particular count of data  we use .loc function

````python
iris.loc[3:50,('SepalLengthCm', 'SepalWidthCm')]
````
